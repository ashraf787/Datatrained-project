# Q2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
html = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
html
```

```python
html.content
```

```python
# Download IMDB's Top rating data
url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
titles = bs.find_all(['h1','h2','h3','h4','h5','h6'])
#all the header tags
print('All Header tags :', *titles, sep='\n\n')
```

```python
# scrap movie names
scraped_movies = soup.find_all('td', class_='titleColumn')
scraped_movies
```

```python
# parse movie names
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n', " ")
    movie = movie.strip(" ")
    k = movie.split(' ')[:-1]
    movie = ' '.join(k[1:])
    movies.append(movie)

movies
```

```python
# scrap rating for movies
scraped_ratings = soup.find_all('td', class_='ratingColumn imdbRating')
scraped_ratings
```

```python
# parse ratings
ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n', '')
    ratings.append(rating)
ratings
```

```python
# scrap  year
scraped_year = soup.find_all('span', class_='secondaryInfo')
scraped_year
```

```python
# parse year
Years = []
for Year in scraped_year:
    Year = Year.get_text().replace(' ','(')
    Year = Year.strip("()")
    Years.append(Year)

Years
```

```python
# Name, IMDB rating, Year of release
data = pd.DataFrame()
data['Movie Name'] = movies[0:100]
data['Rating'] = ratings[0:100]
data['Year of release']= Years[0:100]
data
```
